This piece of code is used to filter to generate the data used for training/testing/validation. We assume we already have a dataset with labels containing the images we needed.

```python
import numpy as np
import pandas as pd
import os
import shutil
import torchvision.transforms as transforms
import PIL
```

```python
#mannually add DataFiltered folder
#see Filter.ipynb in the main directory
#Datafiltered folder contains 12 images with 3 classes 11 of which are transformed using augmentations
#in Datafiltered the plural material classes are removed
```

```python
pd_label = pd.read_csv('labelsPlural.csv')
pd_label
```

```python
#frequency of each class
pd_label['label'].value_counts()
max_count = pd_label['label'].value_counts().max()
max_count
```

```python
source_dir = "./DataFilteredPlural/"
```

```python
#upsample metal
#sample metal
metal_count = pd_label['label'].value_counts()['metal']
metal = pd_label[pd_label['label'] == 'metal']
metal_sample = metal.sample(max_count-metal_count,replace=True)
metal_sample
```

```python
#apply augmentation
for i in range(len(metal_sample)):
    img = PIL.Image.open(source_dir+metal_sample.iloc[i]['filename'])
    img = transforms.RandAugment()(img)
    item_type = metal_sample.iloc[i]['item_type']
    augname = metal_sample.iloc[i]['filename'][:-4]+'_aug' + str(i) + '.png'
    img.save(source_dir+augname)
    #add to csv
    pd_label = pd_label.append({'filename':augname, 'item_type': item_type,'label':'metal'},ignore_index=True)
```

```python
#upsample wood,paper
#sample wood,paper
woodpaper_count = pd_label['label'].value_counts()['wood,paper']
woodpaper = pd_label[pd_label['label'] == 'wood,paper']
woodpaper_sample = woodpaper.sample(max_count-woodpaper_count,replace=True)
woodpaper_sample
```

```python
#apply augmentation
for i in range(len(woodpaper_sample)):
    img = PIL.Image.open(source_dir+woodpaper_sample.iloc[i]['filename'])
    img = transforms.RandAugment()(img)
    item_type = woodpaper_sample.iloc[i]['item_type']
    augname = woodpaper_sample.iloc[i]['filename'][:-4]+'_aug' + str(i) + '.png'
    img.save(source_dir+augname)
    #add to csv
    pd_label = pd_label.append({'filename':augname, 'item_type': item_type,'label':'wood,paper'},ignore_index=True)
```

```python
#save the upsampled csv
pd_label.to_csv('labels_upsampled.csv',index=False)
```
